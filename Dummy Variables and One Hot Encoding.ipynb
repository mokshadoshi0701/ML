```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/moksh/OneDrive/Documents/PythonYoutube/py-master/py-master/ML/5_one_hot_encoding/Exercise/carprices.csv")
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')  #you can also write axis=1
merged
```

```python
final = merged.drop(['Car Model','BMW X5'], axis='columns') #We remove Car Model because its of no use now...but also we remove one of the dummy column to avoid dummy variable trap...and by doing so it won't affect our result!!
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('Sell Price($)', axis = 'columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
model.fit(X,y)
```

```python
model.predict([[45000, 4, 0, 1]])
```

```python
model.predict([[86000, 7, 0, 0]])  #See here we're predicting Price of BMW X5...It doesn't affect if we remove one column because all these are multicorrelated...i.e they can be predicted from one another...menaing 0,0 obviously means BMW X5
```

```python
model.score(X,y)
```

## Linear regression using One Hot Encoding for categorical data

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle['Car Model'] = le.fit_transform(dfle['Car Model'])
dfle
```

```python
X = dfle[['Car Model', 'Mileage', 'Age(yrs)']].values
X
```

```python
y = dfle['Sell Price($)']
y
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([("Car Model", OneHotEncoder(), [0])], remainder = 'passthrough')
X = np.array(ct.fit_transform(X))

```

```python
X
```

```python
X = X[:, 1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0, 1, 45000, 4]])
```

```python
model.predict([[1, 0, 86000, 7]])
```
